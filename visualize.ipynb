# The Base Questionnaire

## Data Import

Import utility class and read data:

```python
import matplotlib.pyplot as plt
import seaborn as sn
import warnings
from questionnaire_reader import QuestionnaireReader
warnings.filterwarnings('ignore')
```

```python
qr = QuestionnaireReader()
data = qr.data
COLORS = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

## Data Acquisition Rate

```python
timestamps = data["Timestamp"].astype("datetime64")
group_by = [timestamps.dt.year, timestamps.dt.month]
collected_by_month = timestamps.groupby(group_by).count()
collected_by_month.name = "Number of Collected Responses by Month"
plot = collected_by_month.plot(kind="bar",
                               figsize=(14, 5),
                               rot=45,
                               title=collected_by_month.name,
                               width=1)
plot.set_xlabel("Year, Month");
plot.set_ylabel("# Collected");
```

```python
_, bfi_corr_ax = plt.subplots(figsize=(14,10))
_ = sn.heatmap(qr.data[BFI_TRAITS].corr(), annot=True, ax=bfi_corr_ax)
```
